New Variables

```python
#Zone-Wise Production - 1997-2014
north_india = ['Jammu and Kashmir', 'Punjab', 'Himachal Pradesh', 'Haryana', 'Uttarakhand', 'Uttar Pradesh', 'Chandigarh']
east_india = ['Bihar', 'Odisha', 'Jharkhand', 'West Bengal']
south_india = ['Andhra Pradesh', 'Karnataka', 'Kerala' ,'Tamil Nadu', 'Telangana']
west_india = ['Rajasthan' , 'Gujarat', 'Goa','Maharashtra']
central_india = ['Madhya Pradesh', 'Chhattisgarh']
north_east_india = ['Assam', 'Sikkim', 'Nagaland', 'Meghalaya', 'Manipur', 'Mizoram', 'Tripura', 'Arunachal Pradesh']
ut_india = ['Andaman and Nicobar Islands', 'Dadra and Nagar Haveli', 'Puducherry']
```

```python
def get_zonal_names(row):
    if row['State Name'].strip() in north_india:
        val = 'North Zone'
    elif row['State Name'].strip()  in south_india:
        val = 'South Zone'
    elif row['State Name'].strip()  in east_india:
        val = 'East Zone'
    elif row['State Name'].strip()  in west_india:
        val = 'West Zone'
    elif row['State Name'].strip()  in central_india:
        val = 'Central Zone'
    elif row['State Name'].strip()  in north_east_india:
        val = 'NE Zone'
    elif row['State Name'].strip()  in ut_india:
        val = 'Union Terr'
    else:
        val = 'No Value'
    return val

df['Zones'] = df.apply(get_zonal_names, axis=1)
df['Zones'].unique()
```

```python
df.Zones.value_counts()
```

```python
data_explore = df.copy()
```

Visualization on Important variables:
  a. Zonal distribution of crops:

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(data_explore.Zones.sort_values(ascending=True), data_explore['Total Production (1000 tons)'])
plt.yscale('log')
plt.title('Zone-Wise Production: Total')
```

Production wise top zone is South India

```python
south_zone =  data_explore[(data_explore["Zones"] == 'South Zone')]
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(south_zone['State Name'], south_zone['Total Production (1000 tons)'],errwidth=0)
plt.yscale('log')
plt.title('Southern-Zone Production')
```

b. Crop wise Production status:

```python
crop = data_explore.groupby(by=df['State Name'])['Total Production (1000 tons)'].sum().reset_index().sort_values(by='Total Production (1000 tons)', ascending=False).head(10)
crop 
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(crop['State Name'], crop['Total Production (1000 tons)'])
plt.yscale('log')
plt.title('Overall Crops vs Production')
```

c. Yearwise Production Status:

```python
plt.tick_params(labelsize=10)
data_explore.groupby("Year")['Total Production (1000 tons)'].agg("sum").plot.bar()
```

d. Season wise Production Status:

```python
#Year vs Production
df_season=data_explore.copy()
season = df_season.groupby(by='Year')['Total Production (1000 tons)'].sum().reset_index().sort_values(by='Total Production (1000 tons)', ascending=False).head(10)
season
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(season.Year, season['Total Production (1000 tons)'],errwidth=0)
plt.yscale('log')
plt.title('Seasonal Crops vs Production')
```

Area-wise crop production:

```python
df = data_explore.groupby(by='State Name')['Total Area (1000 ha)'].sum().reset_index().sort_values(by='Total Area (1000 ha)', ascending=False)
df.head()

fig, ax = plt.subplots(figsize=(25,10))
sns.barplot(df['State Name'].head(10), df['Total Area (1000 ha)'].head(10), errwidth=0)
plt.title('Agricultural Area Distribution - India')
df.head(10)
```

```python
df = data_explore.groupby(by='State Name')['Total Area (1000 ha)'].sum().reset_index().sort_values(by='Total Area (1000 ha)', ascending=False)
df = df.head(5)

fig, ax = plt.subplots(figsize=(25,30), sharey='col')
count = 1

for state in df['State Name'].unique():
    plt.subplot(len(df['State Name'].unique()),1,count)
    sns.lineplot(data_explore[data_explore['State Name']==state]['Year'],data_explore[data_explore['State Name']==state]['Total Production (1000 tons)'], ci=None)
    plt.subplots_adjust(hspace=0.6)
    plt.title(state)
    count+=1;
```
